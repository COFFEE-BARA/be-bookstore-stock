# 교보문고 상세 url

```python
def detail_kyobo(code):
    # 요청 URL
    url = f'https://search.kyobobook.co.kr/search?keyword={code}&gbCode=TOT&target=total'

    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    result_dict = {}

    prod_name_group = soup.find('div', class_='auto_overflow_wrap prod_name_group')

    if prod_name_group:
        a_tag = prod_name_group.find('a', class_='prod_info')
        if a_tag and 'href' in a_tag.attrs:
            href = a_tag['href']
            result_dict[code] = href
        else:
            result_dict[code] = '판매안함'
    else:
        result_dict[code] = '판매안함'

    return result_dict
```

```python
# 영풍문고내에 있는 대표코드와 해당 가격을 가져와야함

def yp_stock(code, price):
    url = f'https://www.ypbooks.co.kr/ypbooks_mobile/sub/mBranchStockLoc.jsp?bookCd={code}&bookCost={price}'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    stores_stock = {}
    store_elements = soup.find_all('td', class_='store')

    for element in store_elements:
        store_name = element.find('strong').text
        stock = element.find('span', class_='stock').text
        stores_stock[store_name] = stock
    
    return stores_stock

check = yp_stock(code, price)
```

# 함수화

```python
def detail_yp(code):
    url = 'https://www.ypbooks.co.kr/ypbooks/search/requestAjaxSearchTab.jsp'
    
    result_dict = {}

    data = {
    'query': code,
    'collection': 'ALL',
    'searchfield': 'ALL',
    'showCnt': '',
    'sortField': 'RANK',
    'notSoldOut': 'Y',
    'catesearch': 'false',
    'c1': '',
    'c2': '',
    'c3': '',
    'viewStyle': 'list',
    'pageNum': '1'
    }

    response = requests.post(url, data=data)
    soup = BeautifulSoup(response.text, 'html.parser')
    checkbox = soup.find('input', {'name': 'checkboxCartBook'})
    cost_span = soup.find('span', {'class': 'cost'})

    checkbox_value = checkbox['value'] if checkbox else None
    cost_text = cost_span.get_text() if cost_span else None
    cost_number = re.sub(r"[^0-9]", "", cost_text) if cost_text else None

    result_dict[code] = [checkbox_value, cost_number] # isbn -> key, value -> [unique_code, price]
    return result_dict
```

```python
check = detail_yp('9788988462096') # 0.1초 걸림
```
